```python
from pyspark.sql import SparkSession
import pandas as pd
```

```python
spark = SparkSession.builder.appName('ml-bank').getOrCreate()
data = spark.read.csv('new.csv', header = True, inferSchema = True)
data.printSchema()
```

```python
pd.DataFrame(data.take(5), columns=data.columns).transpose()
```

```python
from pyspark.ml.feature import OneHotEncoderEstimator, StringIndexer, VectorAssembler
from pyspark.ml import Pipeline
import datetime
import time
```

```python
features = ['Coverage','Education','EmploymentStatus','Gender','Income','Location Code',
            'Marital Status','Monthly Premium Auto','Months Since Last Claim',
            'Months Since Policy Inception','Number of Open Complaints',
            'Number of Policies','Policy Type','Policy','Renew Offer Type',
            'Sales Channel','Total Claim Amount','Vehicle Class','Vehicle Size',
            'State','Customer Lifetime Value']
assembler = VectorAssembler(inputCols=features, outputCol="features")
stages = [assembler]
label_stringIdx = StringIndexer(inputCol = 'Response', outputCol = 'label')
stages += [label_stringIdx]
```

```python
pipeline = Pipeline(stages = stages)
pipelineModel = pipeline.fit(data)
data = pipelineModel.transform(data)
selectedCols = ['label', 'features'] + cols
data = data.select(selectedCols)
data.printSchema()
```

```python
train_list = []
test_list = []
train, test = data.randomSplit([0.1, 0.9], seed = 2018)
train_list.append(train)
test_list.append(test)
train, test = data.randomSplit([0.2, 0.8], seed = 2018)
train_list.append(train)
test_list.append(test)
train, test = data.randomSplit([0.3, 0.7], seed = 2018)
train_list.append(train)
test_list.append(test)
train, test = data.randomSplit([0.4, 0.6], seed = 2018)
train_list.append(train)
test_list.append(test)
train, test = data.randomSplit([0.5, 0.5], seed = 2018)
train_list.append(train)
test_list.append(test)
train, test = data.randomSplit([0.6, 0.4], seed = 2018)
train_list.append(train)
test_list.append(test)
train, test = data.randomSplit([0.7, 0.3], seed = 2018)
train_list.append(train)
test_list.append(test)
train, test = data.randomSplit([0.8, 0.2], seed = 2018)
train_list.append(train)
test_list.append(test)
train, test = data.randomSplit([0.9, 0.1], seed = 2018)
train_list.append(train)
test_list.append(test)
print("Training Dataset Count: " + str(train.count()))
print("Test Dataset Count: " + str(test.count()))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```python
from pyspark.ml.classification import LogisticRegression

logistic_time = []
accuracy_list_logistic=[]
roc_list_logistic=[]
for i in range(len(train_list)):
    print('-------------------------------------------')
    print('train_length:%i \ntest_length:%i'%(train_list[i].count(),test_list[i].count()))


    lr = LogisticRegression(featuresCol = 'features', labelCol = 'label', maxIter=10)
    
    starttime = time.time()

    lrModel = lr.fit(train_list[i])

    endtime = time.time()

    trainingSummary = lrModel.summary
    
    roc = trainingSummary.roc.toPandas()
    predictions = lrModel.transform(test_list[i])
    predictions.select( 'label', 'prediction', 'probability').show(10)

    evaluator = BinaryClassificationEvaluator()
    print('Test Area Under ROC', evaluator.evaluate(predictions))
    roc_list_logistic.append(evaluator.evaluate(predictions))

    evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction",
                                                  metricName="accuracy")
    accuracy = evaluator.evaluate(predictions)
    accuracy_list_logistic.append(accuracy)
    print("Test set accuracy = " + str(accuracy))

#     endtime = time.time()
    # print((endtime - starttime).seconds)
    print('run time: %ss'%(endtime - starttime))
    logistic_time.append(endtime - starttime)
```

```python
print(logistic_time)
plt.plot(logistic_time)
```

```python
plt.plot(roc['FPR'],roc['TPR'])
plt.ylabel('False Positive Rate')
plt.xlabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
print('Training set areaUnderROC: ' + str(trainingSummary.areaUnderROC))
```

```python
from pyspark.ml.classification import DecisionTreeClassifier

decisiontree_time = []
accuracy_list_dt=[]
roc_list_dt=[]
for i in range(len(train_list)):
    print('-------------------------------------------')
    print('train_length:%i \ntest_length:%i'%(train_list[i].count(),test_list[i].count()))
    
    dt = DecisionTreeClassifier(featuresCol = 'features', labelCol = 'label', maxDepth = 3)
    
    starttime = time.time()
    
    dtModel = dt.fit(train_list[i])
    
    endtime = time.time()
    predictions = dtModel.transform(test_list[i])
    predictions.select('label', 'prediction', 'probability').show(10)
    
    evaluator = BinaryClassificationEvaluator()
    print("Test Area Under ROC: " + str(evaluator.evaluate(predictions, {evaluator.metricName: "areaUnderROC"})))
    roc_list_dt.append(evaluator.evaluate(predictions, {evaluator.metricName: "areaUnderROC"}))

    
    evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction",
                                                  metricName="accuracy")
    accuracy = evaluator.evaluate(predictions)
    print("Test set accuracy = " + str(accuracy))
    accuracy_list_dt.append(accuracy)
    
#     endtime = time.time()
    # print((endtime - starttime).seconds)
    print('run time: %s'%(endtime - starttime))
    decisiontree_time.append(endtime-starttime)
```
